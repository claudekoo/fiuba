# Ejercicio 31
Listar la cantidad de reviews neutras según el año de última actualización de la aplicación. Es decir, si una aplicación tiene 10 reviews neutras y fue actualizada por última vez en 2012, se cuentan 10 reviews para ese año. Además listar el nombre de las aplicaciones cuyo año de última actualización corresponde al año de menor cantidad de reviews neutras. (⭐⭐)

```python
neutralAppReviews = validAppReviews[validAppReviews['Sentiment']=='Neutral']
neutralAppReviews
```

```python
neutralReviewsPerApp = neutralAppReviews.groupby(['App']).size().reset_index(name='Neutral Reviews')
neutralReviewsPerApp
```

```python
appLastUpdated = validAppDetails.loc[:, ['App','Last Updated']]
appLastUpdated
```

```python
appLastUpdated['Last Update Year'] = appLastUpdated['Last Updated'].str[-4:].astype('int')
appLastUpdatedYear = appLastUpdated.loc[:, ['App', 'Last Update Year']]
appLastUpdatedYear
```

Pueden existir años donde la mayor cantidad de reviews neutrales es 0, por lo que no se debe utilizar how='inner'(el default), sino tomar de referencia el df con todas las apps y rellenar NaNs con 0. Otra forma es utilizar how='inner' y luego llenar a mano los años que faltan(si es que los hay) con cantidad de reviews neutros 0.

```python
neutralReviewsPerAppWithUpdateYear = pd.merge(appLastUpdatedYear, neutralReviewsPerApp, how='left')
neutralReviewsPerAppWithUpdateYear['Neutral Reviews'] = neutralReviewsPerAppWithUpdateYear['Neutral Reviews'].fillna(0).astype('int')
neutralReviewsPerAppWithUpdateYear
```

La cantidad de reviews neutras según el año de última actualización de la aplicación:

```python
neutralReviewsPerUpdateYear = neutralReviewsPerAppWithUpdateYear.groupby(['Last Update Year'])['Neutral Reviews'].sum()
neutralReviewsPerUpdateYear
```

Lista de las aplicaciones cuyo año de última actualización corresponde al año de menor cantidad de reviews neutras:

```python

yearWithLeastNeutralReviews = neutralReviewsPerUpdateYear.idxmin()
appLastUpdatedYear[appLastUpdatedYear['Last Update Year']==yearWithLeastNeutralReviews]
```
