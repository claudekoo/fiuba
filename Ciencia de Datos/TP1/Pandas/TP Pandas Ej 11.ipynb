Descartamos los nulls en Sentiment Subjectivity

```python
validAppReviews = appReviews[appReviews['Sentiment_Subjectivity'].notna()]
validAppReviews
```

```python
validAppReviews['App']=validAppReviews['App'].astype('string')
validAppReviews['Sentiment_Subjectivity']=validAppReviews['Sentiment_Subjectivity'].astype('float')
validAppReviews.info()
```

# Ejercicio 11
¿Cuál es la aplicación con mayor promedio de score de sentimiento subjetivo? (⭐)

```python
subjectivityMeanPerApp = pd.pivot_table(validAppReviews,values='Sentiment_Subjectivity',index=['App'],aggfunc='mean').reset_index()
subjectivityMeanPerApp
```

La aplicación con mayor promedio de sentimiento subjetivo es Google Slides, con 0.92 de dicho índice:

```python
subjectivityMeanPerApp.loc[subjectivityMeanPerApp['Sentiment_Subjectivity'].idxmax()]
```
